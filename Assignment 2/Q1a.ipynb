```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Reading the tips dataset from csv
data = pd.read_csv('../datasets/tips.csv')
data.head()
```

```python
# Extracting the columns in the specified sequence
selected_columns = ['time', 'total_bill', 'tip', 'day']
data = data[selected_columns]
```

```python
# Setting seaborn style
sns.set(style="whitegrid")
```

```python
# Plotting histogram for total_bill
plt.figure(figsize=(10, 6))
sns.histplot(data['total_bill'], bins=40, kde=False, color='skyblue', edgecolor='black')
plt.title('Histogram of Total Bill')
plt.xlabel('Total Bill')
plt.ylabel('Frequency')
plt.show()
```

We can observe from the histogram plot that the range of total_bill that has maximum frequency is 10-20.

```python
# Plotting bar chart for day
plt.figure(figsize=(8, 6))
sns.countplot(x='day', data=data, palette='bright', hue='day', legend=False)
plt.title('Bar Chart of Day')
plt.xlabel('Day')
plt.ylabel('Count')
```

From the bar chart, We can see that the staturday has the maximum count of total_bill.
