```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Reading the iris dataset
iris = pd.read_csv('iris.csv')
iris.head()
```

```python
# Seaaborn Style
sns.set(style="whitegrid")
```

## Box Plots

```python
# Creating a single figure with subplots
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(12, 10))

# Drawing box plots for each column
sns.boxplot(x='Flowers', y='sepallength', data=iris, ax=axes[0, 0], palette="pastel", hue="Flowers")
axes[0, 0].set_title('Sepal Length')

sns.boxplot(x='Flowers', y='sepalwidth', data=iris, ax=axes[0, 1], palette="pastel", hue="Flowers")
axes[0, 1].set_title('Sepal Width')

sns.boxplot(x='Flowers', y='petallength', data=iris, ax=axes[1, 0], palette="pastel", hue="Flowers")
axes[1, 0].set_title('Petal Length')

sns.boxplot(x='Flowers', y='petalwidth', data=iris, ax=axes[1, 1], palette="pastel", hue="Flowers")
axes[1, 1].set_title('Petal Width')

# Adjusting layout
plt.tight_layout()

# Showing the plots
plt.show()
```

## Histograms

```python
# Creating a single figure with subplots
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(12, 10))

# Drawing Histograms for each column
sns.histplot(iris['sepallength'], ax=axes[0, 0], color="skyblue")
axes[0, 0].set_title('Sepal Length')

sns.histplot(iris['sepalwidth'], ax=axes[0, 1], color="olive")
axes[0, 1].set_title('Sepal Width')

sns.histplot(iris['petallength'], ax=axes[1, 0], color="gold")
axes[1, 0].set_title('Petal Length')

sns.histplot(iris['petalwidth'], ax=axes[1, 1], color="teal")
axes[1, 1].set_title('Petal Width')

# Adjusting layout
plt.tight_layout()

# Showing the plots
plt.show()
```

## Bar Chart to show the number of each type of flower

```python
# Bar Chart
sns.countplot(x='Flowers', data=iris, palette='pastel', hue='Flowers')
```
